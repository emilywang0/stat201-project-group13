H_a: p_1 - p_2 \neq 0
$$

### Preliminary exploratory data analysis

To get an idea of the distribution of the crimes committed in both Kerrisdale and Strathcona, we first filter and count by `TYPE` and `YEAR` to get a tally for every crime committed, then we plot the raw data in a stacked bar plot.

```r
# Filter for crimes committed in Kerrisdale
kerrisdale_crimes <- crime_mutate |>
    filter(NEIGHBOURHOOD == "Kerrisdale")

# Filter for crimes committed in Strathcona
strathcona_crimes <- crime_mutate |>
    filter(NEIGHBOURHOOD == "Strathcona")

# Get counts of Kerrisdale crimes
kerrisdale_crime_counts <- kerrisdale_crimes |>
    count(TYPE, YEAR)

# Get counts of Strathcona crimes
strathcona_crime_counts <- strathcona_crimes |>
    count(TYPE, YEAR)

# Preview data frames
print("Table 3: Kerrisdale Crime Counts")
head(kerrisdale_crime_counts)

print("Table 4: Strathcona Crime Counts")
head(strathcona_crime_counts)
```

```r
# Create stacked bar plot of raw data
kerrisdale_raw_data_plot <- kerrisdale_crime_counts |>
    ggplot(aes(x = YEAR, y = n, fill = TYPE)) +
    geom_bar(stat = "identity") +
    labs(x = "Year", y = "Number of Crimes", fill = "Type of Crime", caption = "Figure 1: Stacked Bar Plot of Crimes in Kerrisdale") +
    ggtitle("Crimes in Kerrisdale by Type and Year") +
    ylim(0, 2500) + theme(plot.caption = element_text(hjust = 0.5))

# Create stacked bar plot of raw data
strathcona_raw_data_plot <- strathcona_crime_counts |>
    ggplot(aes(x = YEAR, y = n, fill = TYPE)) +
    geom_bar(stat = "identity") +
    labs(x = "Year", y = "Number of Crimes", fill = "Type of Crime", caption = "Figure 2: Stacked Bar Plot of Crimes in Strathcona") +
    ggtitle("Crimes in Strathcona by Type and Year") +
    ylim(0, 2500) + theme(plot.caption = element_text(hjust = 0.5))

# Display plots
grid.arrange(kerrisdale_raw_data_plot, strathcona_raw_data_plot)
```

Given these two plots, it is clear that Strathcona sees a greater total number of crimes. However, it is difficult to judge which neighbourhood has a higher proportion of thefts. Thus, to calculate the proportion of thefts, we group the data into thefts and non-thefts. Our new column, 'Theft' includes the following theft-related crimes: "Theft of Bicycle", "Theft of Vehicle", "Theft from Vehicle", and "Other Theft". Non-theft related crimes are under the column 'Other', which includes crimes such as "Break and Enter Commercial", "Homicide", among others. We can then calculate the total number of theft related crimes and non-theft related crimes.

```r
# Group theft related crimes under Theft and non-theft related crimes under Other
kerrisdale_crimes_grouped <- kerrisdale_crimes |>
    mutate(TYPE = ifelse(TYPE %in% c("Theft of Bicycle", "Theft of Vehicle", "Theft from Vehicle", "Other Theft"), 
                         "Theft", "Other")) 

strathcona_crimes_grouped <- strathcona_crimes |>
    mutate(TYPE = ifelse(TYPE %in% c("Theft of Bicycle", "Theft of Vehicle", "Theft from Vehicle", "Other Theft"), 
                         "Theft", "Other"))

# Get total number of theft and non theft crimes and find the proportion 
kerrisdale_counts <- kerrisdale_crimes_grouped |>
    count(TYPE) |> mutate(prop = n / sum(n), label = scales::percent(prop))

strathcona_counts <- strathcona_crimes_grouped |>
    count(TYPE) |> mutate(prop = n / sum(n), label = scales::percent(prop))
```

To get a visual understanding of the proportions for each neighbourhood, we will make a pie chart of the total theft vs non-theft crimes.

It is clear that the overall proportions are similar between the two neighbourhoods, but before we reach a conclusion, we can obtain the proportion of theft in Kerrisdale and Strathcona in each individual year to get a better idea of the variation.

```r
# Find proportion of theft for Kerrisdale by year
kerrisdale_yearly_proportions <- kerrisdale_crimes_grouped |>
    group_by(YEAR, TYPE) |>
    summarise(count = n()) |>
    spread(TYPE, count, fill = 0) |>
    mutate(Proportion = Theft / (Theft + Other))

# Find proportion of theft for Strathcona by year
strathcona_yearly_proportions <- strathcona_crimes_grouped |>
    group_by(YEAR, TYPE) |>
    summarise(count = n()) |>
    spread(TYPE, count, fill = 0) |>
    mutate(Proportion = Theft / (Theft + Other))

print("Table 6: Kerrisdale Proportion of Theft by Year")
kerrisdale_yearly_proportions

print("Table 7: Strathcona Proportion of Theft by Year")
strathcona_yearly_proportions
```

### Parameter Estimates

### Results

#### Bootstrapping Method
First, we will use bootstrapping to conduct a hypothesis test using the null distribution. We will then obtain a confidence interval using both the percentile and standard error method. 
##### Hypothesis Testing using Bootstrapping
We can test $H_0$ by first assuming it to be true and constructing the null distribution. We will then compute the observed test statistic, which is the observed difference in proportion of theft between Kerrisdale and Strathcona.

```r
# First, combine the two data frames by row so we can do data analysis using the infer package
crimes_grouped <- rbind(kerrisdale_crimes_grouped, strathcona_crimes_grouped)

# Calculate the null distribution
null_distribution <- crimes_grouped |>
    specify(formula = TYPE ~ NEIGHBOURHOOD, success = "Theft") |>
    hypothesize(null = "independence") |>
    generate(reps = 100, type = "permute") |>
    calculate(stat = "diff in props", order = c("Kerrisdale", "Strathcona"))

# Find the observed test statistic
obs_diff_prop <- crimes_grouped |>
    specify(TYPE ~ NEIGHBOURHOOD, success = "Theft") |>
    calculate(stat = "diff in props", order = c("Kerrisdale", "Strathcona"))

print("Observed test statistic:")
obs_diff_prop
```

Now, we can visualize the null distribution with our p-value represented as the red shaded region. The solid dark red line is our observed test statistic.

```r
visualize(null_distribution) +
    shade_p_value(obs_stat = obs_diff_prop, direction = "both") +
    labs(x = "Difference in Proportion", caption = "Figure 5: Simulation-Based Null Distribution") +
    theme(plot.caption = element_text(hjust = 0.5))
```

Judging by the shaded region in Figure 5, we can see that 0.0019, which is the observed difference of proportion is quite common under our null hypothesis. In other words, since the shaded region is large, our p-value will be large. It seems as if we will fail to reject $H_0$ Let's verify this by obtaining the exact value of our p-value.

```r
null_distribution |>
    get_p_value(obs_stat = obs_diff_prop, direction = "both")
```

Since the p-value 0.66 is greater than the significance level $α = 0.05$, we fail to reject $H_0$. We do not have enough evidence to prove a statiscially significant difference in proportion of crime between Kerrisdale and Strathcona. 

Because we fail to reject $H_0$, we are at risk of making a Type II error, which would be failing to conclude that the difference in proportions is significant when it in fact is.

#### Confidence Intervals using Bootstrap Distribution
Now, let's obtain a 95% confidence interval for the difference in proportion. To do so, we will need to use the bootstrap distribution.

```r
bootstrap_distribution <- crimes_grouped |>
    specify(formula = TYPE ~ NEIGHBOURHOOD, success = "Theft") |>
    generate(reps = 100, type = "bootstrap") |>
    calculate(stat = "diff in props", order = c("Kerrisdale", "Strathcona"))
```

Let's first compute a confidence interval using the percentile method.

```r
percentile_ci <- bootstrap_distribution |>
    get_confidence_interval(level = 0.95, type = "percentile")

percentile_ci
```

The shorthand interpretation for the 95% confidence interval would be as follows: we are 95% confident that the true difference in population proportions of Kerrisdale and Strathcona is between (-0.0074, 0.0111). Notice how 0 is included in this confidence interval.

Now let's visualize the bootstrap distribution with the percentile-based confidence interval shaded in.

```r
visualize(bootstrap_distribution) + 
    shade_confidence_interval(endpoints = percentile_ci) +
    labs(x = "Difference in Proportion", caption = "Figure 6: Percentile-based 95% confidence interval") +
    theme(plot.caption = element_text(hjust = 0.5))
```

From Figure 6, we can observe that our bootstrap distribution appears to be roughly normally shaped. Hence, we can also use the standard error method to obtain the 95% confidence interval. To do this, we must also include the observed difference in proportion as the center of the confidence interval.

```r
se_ci <- bootstrap_distribution |>
    get_confidence_interval(level = 0.95, type = "se", point_estimate = obs_diff_prop)

se_ci
```

The shorthand interpretation for the 95% confidence interval would be as follows: we are 95% confident that the true difference in population proportions of Kerrisdale and Strathcona is between (-0.0083, 0.0121). Compared to the percentile-based confidence interval, the standard error-based one is slightly wider. Notice how 0 is also included in this confidence interval.

We can then revisualize the bootstrap distribution with the standard error-based 95% confidence interval.

```r
visualize(bootstrap_distribution) + 
    shade_confidence_interval(endpoints = se_ci) +
    labs(x = "Difference in Proportion", caption = "Figure 7: Standard error-based 95% confidence interval") +
    theme(plot.caption = element_text(hjust = 0.5))
```

#### Asymptotic Method
We will now conduct a hypothesis test and obtain the confidence interval using an asymptotic method with the intention to compare both bootstrapping and asymptotic methods. Asymptotics requires quite a few assumptions to construct the null distribution and is an approximation to the simulation-based method of bootstrapping. By comparing the results of the two methods, we will see if the assumptions for asymptotics hold true for our dataset.

##### Hypothesis Testing using Asymptotics
The assumptions for a hypothesis test for the difference of two population proportions is as follows:
- The size of sample must be large enough
- The sample must be taken in independent fashion
- The sample is random

To test $H_0$, we use the following test statistic, $Z$:

$$
Z = \frac{\hat{p}_1 - \hat{p}_2}{\sqrt{\hat{p}(1-\hat{p})\left(\frac{1}{n_1} + \frac{1}{n_2}\right)}}
$$
where $\hat{p}_1$ and $\hat{p}_2$ are the sample proportions in Kerrisdale and Strathcona, respectively; $n$ is the sample size; and $\hat{p}$ is the pooled proportion, given by: 

$$\hat{p} = \frac{n_1\hat{p}_1+n_2\hat{p}_2}{n_1+n_2}$$

Since the population distribution is clearly not Normal since the random variable is a 0-1 variable, we need to rely on the Central Limit Theorem (CLT). The CLT states that given that the sample size is large enough, the null model of $Z$ is approximately $N(0, 1)$.

First, we will calculate the test statistic.

```r
# Calculate sample size for Kerrisdale and Strathcona 
n1 <- sum(crimes_grouped$NEIGHBOURHOOD == "Kerrisdale")
n2 <- sum(crimes_grouped$NEIGHBOURHOOD == "Strathcona")

# Calculate sample proportions in Kerrisdale and Strathcona
phat1 <- crimes_grouped |> filter(NEIGHBOURHOOD == "Kerrisdale", TYPE == "Theft") |>
                                        summarize(phat = n()/n1) |> pull()
phat2 <- crimes_grouped |> filter(NEIGHBOURHOOD == "Strathcona", TYPE == "Theft") |>
                                        summarize(phat = n()/n2) |> pull()

# Calculate pooled proportion
p <- (n1*phat1+n2*phat2)/(n1 + n2)

# Finally, calculate test statistic
test_stat <- (phat1-phat2)/sqrt(p*(1-p)*(1/n1+1/n2))

print("Test statistic:")
test_stat
```

Now, we will calculate the p-value using `pnorm`.

```r
 p_val <- 2 * pnorm(test_stat, lower.tail = FALSE)

print("p-value:")
p_val
```

We can also verify this value by using `prop.test`.

```r
# Calculate number of successses in each sample
x1 <- crimes_grouped |> filter(NEIGHBOURHOOD == "Kerrisdale", TYPE == "Theft") |> nrow()
x2 <- crimes_grouped |> filter(NEIGHBOURHOOD == "Strathcona", TYPE == "Theft") |> nrow()

# Find p-value
prop_test <- tidy(
        prop.test(x = c(x1,x2), 
                  n = c(n1,n2), 
                  alternative = "two.sided",
                  correct = FALSE))     

print("Results of prop.test")
prop_test
```

Since the p-value 0.71 is greater than the significance level $α = 0.05$, we fail to reject $H_0$. We do not have enough evidence to prove a statiscially significant difference in proportion of crime between Kerrisdale and Strathcona. 

Recall that the p-value for our bootstrapping method was 0.66 and compare that with our p-value of 0.71 using asymptotics. The p-values have a difference of 5% with asymptotics' p-value being higher than bootstrapping's p-value. Despite this difference, both p-values lead to the same conclusion - a failure to reject the null hypothesis. Thus, it seems that the assumptions  are reasonable to make and asymptotics is a suitable method for this hypothesis test.

This suggests that regardless of the method used (bootstrapping or asymptotics), the evidence from the data does not provide sufficient support to refute the null hypothesis in favor of the alternative hypothesis. In other words, there isn't enough statistical significance to claim a difference in theft proportions based on the observed data.

#### Confidence Intervals using Asymptotics
Now, let's obtain a 95% confidence interval for the difference in proportion. To do so using asymptotics, we will use the following formulas:

$$CI({p}_1 - {p}_2) = (\hat{p}_1 - \hat{p}_2) \pm z^* \sqrt{\frac{\hat{p}_1(1-\hat{p}_1)}{n_1}+\frac{\hat{p}_2(1-\hat{p}_2)}{n_2}}$$

```r
# Obtain the critical value
crit_value <- qnorm(1-0.05/2)

# Calculate the confidence interval using the formula
crime_ci <- tibble(
     lower_ci = (phat1 - phat2) - crit_value * sqrt(phat1*(1-phat1)/n1 + phat2*(1-phat2)/n2),
     upper_ci = (phat1 - phat2) + crit_value * sqrt(phat1*(1-phat1)/n1 + phat2*(1-phat2)/n2)
 )

crime_ci
```

The shorthand interpretation for the 95% confidence interval would be as follows: we are 95% confident that the true difference in population proportions of Kerrisdale and Strathcona is between (-0.0081, 0.0119). Notice that 0 also lies within the interval. Now, we can compare the asymptotic confidence interval with the two bootstrapping ones.

```r
# Combine all 3 confidence intervals into one tibble to compare
combined_ci <- bind_rows(
  percentile_ci %>% mutate(type = "percentile"),
  se_ci %>% mutate(type = "se"),
  crime_ci %>% mutate(type = "asymptotic")
) %>%
select(type, lower_ci, upper_ci)

combined_ci
```

We observe that the asymptotic confidence interval has the widest range, followed by standard error-based and percentile-based has the smallest range. However, the difference in range is minimal and the confidence intervals are all extremely similar. They also all include the value 0.
